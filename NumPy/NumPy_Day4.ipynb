# 3D plotting

```python
import numpy as np
import pandas as pd

a = np.linspace(-10, 9, 21, dtype=int)
b = np.linspace(-10, 9, 21, dtype=int)

xx, yy = np.meshgrid(a, b)

print(xx)
```

- Plotting 3D graphs
- meshgrid

```python
import matplotlib.pyplot as plt
plt.scatter(xx,yy)
```

```python
def fun(x,y):
    return 3*np.sin(x) + 2*y
zz=fun(xx,yy)
```

```python
# import 3D library
import plotly.express as px
import plotly.graph_objects as go
```

```python
fig = px.scatter_3d()

fig.add_trace(go.Surface(x=xx,y=yy,z=zz))

fig.show()
```

### Working with images

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
img = cv2.imread('broadcasting-rules.png')

# image show
plt.imshow(img)
```

```python
# flip()
flip_img=np.flip(img)
plt.imshow(flip_img)
```

```python
# clip()
plt.imshow(np.clip(img,8,200))
```

```python
plt.imshow(255-img)
```

```python
plt.imshow(img[100:500,100:900,:])
```

```python
# Plot Histogram() flatten is use to convert 2d array to 3d array
plt.hist(img.flatten(),bins=255)
```
