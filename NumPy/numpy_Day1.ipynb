## Numpy Arrays Vs Sequence
- Python sequence stores mixed data types; NumPy array usually stores same data type.
- Python sequence is slower; NumPy array is faster for large data.
- Python sequence uses more memory; NumPy array uses less memory.
- Python sequence needs loops for math; NumPy array does math directly.
- Python list * 2 repeats data; NumPy array * 2 multiplies each element.
- Python sequence is good for normal coding; NumPy array is best for data science, ML, and numerical

# Array Functions
- Min / Max / Sum / Product

```python
a1=np.random.random((3,3))
a1=np.round(a1*100)
a1
```

```python
# min/ max / sum / product 
a1.astype(np.int32)
print(f"Minimum value: {np.min(a1)}")
print(f"Maximum value: {np.max(a1)}")
print(f"Sum: {np.sum(a1)}")
print(f"Product: {np.prod(a1)}")
```

```python
#axis in max/ min / sum / product
np.max(a1, axis=0) # column wise max
a1.min(axis=1) # row wise min
```

- mean / Median / Std / Var

```python
print(np.mean(a1,axis=1)) # mean of all elements
print(np.median(a1,axis=1)) # median of all elements in each row
print(np.std(a1,axis=1)) # standard deviation of all elements in each row
print(np.var(a1,axis=1)) # variance of all elements in each row

# only need first row mean, median, std, var
print(np.mean(a1[0])) # mean of first row
# only need first column mean, median, std, var
print(np.mean(a1[:,0])) # mean of first column
```

- trignomatric function

```python
np.sin(a1)
```

- dot product

```python
# dot product
a1=np.array([[1,2],[3,4]])
a2=np.array([[5,6],[7,8]])
np.dot(a1,a2)
```

- Log and Exponents

- Round / Floor / Ceil

```python
#round: round the elements in the array to the nearest integer
np.round(np.random.random((2,2))*100)
```

```python
# floor: round the elements in the array to the nearest integer towards negative infinity
np.floor(np.random.random((2,2))*100)
```

```python
# ceil: round the elements in the array to the nearest integer towards positive infinity
np.ceil(np.random.random((2,2))*100)
```
