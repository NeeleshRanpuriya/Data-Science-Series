### np.concatenate()
- https://numpy.org/doc/stable/reference/generated/numpy.concatenate.html

```python
# np.concatenate()
c=np.arange(6).reshape(2,3)
d=np.arange(6,12).reshape(2,3)

print(c)
print(d)
```

```python
np.concatenate((c,d),axis=0) # row wise concatenation
```

```python
np.concatenate((c,d),axis=1) # column wise concatenation
```

### Set Funtions()
- np.union1d
- np.intersect1d
- np.setdiff1d
- np.setxor1d
- np.in1d

```python
# union in 1d 
m=np.array([1,2,3,4,5])
n=np.array([3,4,5,6,7])
np.union1d(m,n)
```

```python
# intersection in 1d
np.intersect1d(m,n)
```

```python
# symmetric difference in 1d
np.setdiff1d(m,n)
```

```python
# set xor symmetric difference in 1d
np.setxor1d(m,n)
```
