Q-2 Ask user to input two numbers a, b. Write a program to generate a random array of shape (a, b) and print the array and avg of the array.

```python
#Code here Q-2 Ask user to input two numbers a, b. Write a program to generate a random array of shape (a, b) and print the array and avg of the array.
a=int(input("Enter the number of rows: "))
b=int(input("Enter the number of columns: "))
random_array = np.random.random((a, b))
print("Random Array:")
print(random_array)
average = np.mean(random_array)
print("Average:", average)
```

Q-3Write a function to create a 2d array with 1 on the border and 0 inside. Take 2-D array shape as (a,b) as parameter to function.
Eg.-

[[1,1,1,1],<br>
[1,0,0,1],<br>
[1,0,0,1],<br>
[1,1,1,1]]

```python
def boarder(a,b):
    arr=np.ones((a,b))
    arr[1:-1,1:-1]=0
    return arr
boarder(5,3)
```
