```python
def remove_puncuation(text):
    for i in string.punctuation:
        if i in text:
            text=text.replace(i,'')
    return text
```

```python
# Tokanixation: Word level/ Sentance lavel
from nltk.tokenize import word_tokenize
```

```python
import nltk
nltk.download('punkt_tab')

# Tokenization: Word level / Sentence level
df['word_tokens'] = df['review'].apply(word_tokenize)
df.head()
```

```python
import nltk
nltk.download('stopwords')

# Stop Word Removal: is the on
from nltk.corpus import stopwords
stopwords.words('english')
```

```python
def remove_stopword(text):
    L=[]
    for word in text:
        if word not in stopwords.words('english'):
            L.append(word)

    return L
```

```python
df['word_tokens'] = df['word_tokens'].apply(remove_stopword)
```
