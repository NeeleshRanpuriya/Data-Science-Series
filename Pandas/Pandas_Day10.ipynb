### Timedelta Object
Represent a duration, the diffrence between two dates or times.

```python
# create using timestamp objects
t1=pd.Timestamp('17th Nov 2005 12:45:10')
t2=pd.Timestamp('17th Nov 2026 00:00:00')

t2-t1
```

```python
# standalone creation
pd.Timedelta(days=2,hours=10,minutes=25,seconds=10,milliseconds=24,microseconds=28)
```

```python
# Airthmatic Operations
pd.Timestamp('17th Nov 2026 00:00:00')+pd.Timedelta(days=2,hours=10,minutes=25)
```

```python
pd.date_range(start='2005-11-17',end='2026-11-17', freq='D')+pd.Timedelta(days=2,hours=10,minutes=25)
```

```python
# Real life example
df=pd.read_csv('Datasets/deliveries copy.csv')
df.head()
```

```python
# Average Delivery Time
df['delivery_date']=pd.to_datetime(df['delivery_date'])
df['order_date']=pd.to_datetime(df['order_date'])
```
