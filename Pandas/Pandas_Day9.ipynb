### Datetime Index Object
A collection of Pandas timestamp

```python
# from String
pd.DatetimeIndex(['2023/1/1','2022/1/1','2021/1/1'])
```

```python
# using Python datetime object
pd.DatetimeIndex([dt.datetime(2023,12,17),dt.datetime(2023,12,18)])
```

```python
# Using pandas pd.timestamp
dt_index=pd.DatetimeIndex([pd.Timestamp(2023,12,17),pd.Timestamp(2023,12,18)])
```

```python
# using datetimeindex a a series index
pd.Series([1,2],index=dt_index)
```

### Date_range function

```python
# D->Generate daily dates in a given range
pd.date_range(start='2026-06-03',end='2026-06-25',freq='D')
```

```python
# 2D/3D/4D ->skip dates
pd.date_range(start='2026-06-03',end='2026-06-25',freq='2D')
```

```python
# B ->only business days skips saturday sunday
pd.date_range(start='2026-06-03',end='2026-06-25',freq='B')
```

```python
# W -> one weak per day
pd.date_range(start='2026-06-03',end='2026-06-25',freq='w-mon')
```

```python
# H-> hourly date(factor)
pd.date_range(start='2026-06-03',end='2026-06-25',freq='24H')
```

```python
# m -> month end
pd.date_range(start='2026-06-03',end='2026-06-25',freq='M')
```

```python
# ms -> month start
pd.date_range(start='2026-06-03',end='2026-06-25',freq='MS')
```

```python
# Using periods (Number of results)
pd.date_range(start='2026-06-03',periods=31,freq='D')
```

### to_datetime function

```python
# Simple series Exmaple
s=pd.Series(['2023/1/1','2023/1/2'])
pd.to_datetime(s).dt.day_name()
```

```python
# with errors solve
s=pd.Series(['2023/1/1','2023/100/2'])
pd.to_datetime(s,errors='coerce').dt.month_name()
```
