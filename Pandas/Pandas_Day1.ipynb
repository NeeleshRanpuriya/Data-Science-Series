### What is Pandas
Padas is a fast, powerful, flexible and easy to open source data analysis and manipulation tool, build on a top of the python programming language

Pandas Series<br>
A pandas Series is like a column in a table, it is a 1-D array holding data of any type.

### importing Pandas

```python
import pandas as pd
import numpy as np
```

### Series from list

```python
# string
country=['india','pakistan','USA','Srilanka','Bangladesh']
pd.Series(country)
```

```python
#integer Series
run=[14,45,100,92,108]
run=pd.Series(run)
print(run)
```

```python
# custom index
marks=[25,30,45,50,49]
Subjects=['Maths','Physics','Chemistry','English','Hindi']
pd.Series(marks, index=Subjects)
```

```python
# Setting a name
pd.Series(marks, index=Subjects , name='Neelesh ke Marks')
```

### Series from dist

```python
marks={
    'Maths':25,
    'Physics':30,
    'Chemistry':45,
    'English':50,
    'Hindi':49
}
marks_series=pd.Series(marks, name='Neelesh ke Marks')
print(marks_series)
```

```python
a=pd.Series([1,1,1,2,2,6,4,5,6,7,8,9])
a.duplications()
```

```python
numvalues=pd.Series([1,np.nan,3,np.nan,5,np.nan,7])
numvalues.isnull().sum()
```

```python
# remove null values
numvalues.dropna()
```

```python
# replace null values with mean
numvalues.fillna(numvalues.mean())
```
