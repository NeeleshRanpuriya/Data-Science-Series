### Filtring a DataFrame

```python
ipl.head(2)
```

```python
# find all the final winners
marks=ipl['MatchNumber']=='Final'
new_df=ipl[marks]
new_df[['Season','WinningTeam']]
ipl[ipl['MatchNumber']=='Final'][['Season','WinningTeam']]
```

```python
# how many SuperOvers have been played in IPL?
ipl[ipl['SuperOver']=='Y'].shape[0]
```

```python
# How many mathches has csk won in chennai?
ipl[(ipl['City']=='Chennai') & (ipl['WinningTeam']=='Chennai Super Kings')].shape[0]
```

```python
# toss winner match winner in percentage
(ipl[ipl['TossWinner'] == ipl['WinningTeam']].shape[0]/ipl.shape[0])*100
(ipl[ipl['TossWinner'] != ipl['WinningTeam']].shape[0]/ipl.shape[0])*100

```

```python
# movies witg rating higher than 8 and number of votes higher than 100000
movies.head(2)
```

```python
movies[(movies['imdb_rating']>8) & (movies['imdb_votes']>10000)].shape[0]
```

```python
movie=movies['genres'].str.split('|').apply(lambda x: 'Action' in x)
rating=movies['imdb_rating']>8
movies[movie & rating].shape[0]
```

```python
# write a function that can return the track record of two teams against each other in IPL
def count_wins(team1,team2):
    matches = ipl[
        ipl['Team1'].isin([team1, team2]) &
        ipl['Team2'].isin([team1, team2])
    ]

    return matches['WinningTeam'].value_counts()
count_wins('Chennai Super Kings','Mumbai Indians')
# ipl.head(2)
```
