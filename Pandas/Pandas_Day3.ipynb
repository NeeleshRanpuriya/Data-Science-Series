### Rank in Series

Har value ko uski relative position ke hisaab se rank deta hai.

```python
Batsman=pd.read_csv('Datasets/batsman_runs_ipl.csv')
Batsman.head()
```

```python
Batsman['Batsman_rank'] = Batsman['batsman_run'].rank(ascending=False)

top_batsman = Batsman[Batsman['batsman_run'] > 4000].sort_values(
    by='batsman_run',
    ascending=False
)

top_batsman.plot(
    x='batter',
    y='batsman_run',
    kind='line',
    figsize=(12, 6)
)
```

### Apply Function Row-wise in DataFrames

Har row par custom function apply karke new result nikalne ke liye use hota hai.

```python
points_df = pd.DataFrame(
    {
        '1st point':[(3,4),(-6,5),(0,0),(-10,1),(4,5)],
        '2nd point':[(-3,4),(0,0),(2,2),(10,10),(1,1)]
    }
)

points_df
```

```python
# find ecludian distance
def euclidean(row):
  pt_A = row['1st point']
  pt_B = row['2nd point']
  print(type(pt_A))

  return ((pt_A[0] - pt_B[0])**2 + (pt_A[1] - pt_B[1])**2)**0.5
```

```python
points_df['distance'] = points_df.apply(euclidean,axis=1)
points_df
```
