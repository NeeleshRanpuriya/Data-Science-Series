### Seaborn

```python
import seaborn as sns
```

```python
# import Datase
tips=sns.load_dataset('tips')
iris=sns.load_dataset('iris')
```

###  Categorical Plots

#### 1) catagorical Scatter plot

- Strip plot 
- swarm plot

```python
# sns.scatterplot(data=tips,x='total_bill',y='tip')
tips.head()
sns.stripplot(data=tips,x='day',y='total_bill',jitter=False)
```

```python
sns.catplot(data=tips,x='day',y='total_bill',kind='strip',jitter=True)
```

```python
# jitter
sns.catplot(data=tips,x='day',y='total_bill',kind='strip',jitter=0.5 ,color='blue')
```

-  swarmPlot

```python
sns.swarmplot(data=tips,x='day',y='total_bill',color='red')
```

```python
# hue
sns.catplot(data=tips,x='day',y='total_bill',kind='swarm',hue='sex')
```

### 2) Categorical Distribution Plots

#### Boxplot

```python
# Boxplot
sns.boxplot(data=tips,x='sex',y='total_bill',hue='sex')
```

```python
# Using catplot
sns.catplot(data=tips,x='sex',y='total_bill',kind='box',hue='day')
```

#### Violinplot (Boxplot+KDE Plot)

```python
sns.violinplot(data=tips,x='day',y='total_bill',hue='day')
```

```python
sns.catplot(data=tips,x='day',y='total_bill',kind='violin',hue='sex',split=True)
```

### 3) Categorical Estimation Plot -> for Central Tendency

#### Bar plot

```python
sns.barplot(data=tips,x='sex',y='total_bill',hue='sex')
```

```python
import numpy as np
sns.catplot(data=tips,x='day',y='total_bill',kind='bar',hue='day',estimator=np.sum)
```

#### Point plot

```python
import numpy as np
sns.catplot(data=tips,x='sex',y='total_bill',kind='point',hue='smoker')
```

```python
sns.catplot(data=tips,x='sex',y='total_bill',kind='box',col='smoker')
```

### Multigrid Plot

### Faced Plots -> FacetGrid

```python
# Figlevel -> relplot -> displot -> catplot ->lmplot
```

```python
sns.catplot(data=tips,x='sex',y='total_bill',kind='violin',col='day',row='time')
```

```python
g=sns.FacetGrid(data=tips,col='day',row='time',hue='smoker')
g.map(sns.scatterplot, 'sex', 'total_bill')
g.add_legend()
```

#### Pair Plot

```python
iris
sns.pairplot(data=iris,hue='species')
```

#### Pair Grid

```python
g=sns.PairGrid(data=iris,hue='species')
g.map(sns.scatterplot)
g.add_legend()
```

```python
# map_diag -> map_offdiag
g=sns.PairGrid(data=iris,hue='species')
g.map(sns.scatterplot)
```

```python
g=sns.PairGrid(data=iris,hue='species')
g.map_diag(sns.histplot)
g.map_offdiag(sns.kdeplot)
```

```python
g=sns.PairGrid(data=iris,hue='species')
g.map_diag(sns.histplot)
g.map_offdiag(sns.kdeplot)
g.map_upper(sns.scatterplot)
g.map_lower(sns.scatterplot)
```

#### Join plot

```python
sns.jointplot(data=tips,x='total_bill',y='tip',kind='hist',hue='sex')
```

```python
g=sns.JointGrid(data=tips[['total_bill','tip']],x='total_bill',y='tip')
g.plot(sns.scatterplot,sns.histplot)
```
