#### 1)  Open a local Csv file

```python
import numpy as np
import pandas as pd
df = pd.read_csv('file.csv')
df.head()
```

#### 3) Step Perameter

```python
df= pd.read_csv('file.csv', sep='\t')
```

#### 4) Header Parameter

```python
pd.read_csv('file.csv',header=2)
```

#### 5) Index Col

```python
pd.read_csv('file.csv',index_col='name')
```

#### 6) Use Cols Parameter

```python
pd.read_csv('file.csv',usecols=['name'])
```

#### 7) Squeeze parameter

```python
df = pd.read_csv('file.csv')
print(df.shape)
print(df.columns)
```

#### 8) Skip Rows

```python
pd.read_csv('file.csv',skiprows=2, nrows=5)
```

#### 9) Encoding Prameter

```python
pd.read_csv('file.csv',encoding='utf-8')
```

#### 10) Skip bad line

```python
pd.read_csv(
    'file.csv',
    sep=',',
    on_bad_lines='skip'
)
```

#### 11) dtype parameter

```python
pd.read_csv('file.csv', dtype={'age': np.int8, 'name': 'str'})
```

#### 12) Handling Dates

```python
pd.read_csv('file.csv',parse_dates=['age'])
```

#### 13) Convertor

```python
pd.read_csv('file.csv',converters={'name': lambda x: x.upper()})
```

#### 15) na_values Parameter

```python
pd.read_csv('file.csv',na_values=['NA', 'N/A', 'na', 'n/a', 'null', 'NULL','21'])
```

#### 16) Loading a huge dataset in chunks

```python
dfs=pd.read_csv('file.csv',chunksize=2)
for chunk in dfs:
    print(chunk.shape)#######fefd
```
