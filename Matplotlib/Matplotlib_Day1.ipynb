### 2D line plot

- Bivariant Ananlysis
- categorical -> numerical and numerical -> numerical
- Use Case -> Time Seires Data

```python
# Plotting a simple function
price=[24000,44000,40000,20000,30000]
year=[2015,2016,2017,2018,2019]

plt.plot(year,price)
```

```python
import pandas as pd
# load Datasets
batsman=pd.read_csv('Datasets/sharma-kohli.csv')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'])
```

```python
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
```

```python
# lable title
plt.plot(batsman['index'],batsman['V Kohli'],color='red')
plt.plot(batsman['index'],batsman['RG Sharma'],color='blue')

plt.title('Rohit Sharma Vs V Kohli comparision')
plt.xlabel('Season')
plt.ylabel("Runs scores")
```

```python
# lable title
plt.plot(batsman['index'],batsman['V Kohli'],color='red',linestyle='solid',linewidth=3)
plt.plot(batsman['index'],batsman['RG Sharma'],color='blue',linestyle='dashdot')

plt.title('Rohit Sharma Vs V. Kohli comparision')
plt.xlabel('Season')
plt.ylabel("Runs scores")
```

```python
# lable title
plt.plot(batsman['index'],batsman['V Kohli'],color='red',linestyle='solid',linewidth=3,marker='o',markersize=10)
plt.plot(batsman['index'],batsman['RG Sharma'],color='blue',linestyle='dashdot',marker='o')

plt.title('Rohit Sharma Vs V. Kohli comparision')
plt.xlabel('Season')
plt.ylabel("Runs scores")
```

```python
# Legend -> location
# lable title
plt.plot(batsman['index'],batsman['V Kohli'],color='red',linestyle='solid',linewidth=3,marker='o',markersize=10,label='V.kohli')
plt.plot(batsman['index'],batsman['RG Sharma'],color='blue',linestyle='dashdot',marker='o',label='Rohit Sharma')

plt.title('Rohit Sharma Vs V. Kohli comparision')
plt.xlabel('Season')
plt.ylabel("Runs scores")
plt.legend()
```

```python
# limiting axes
price=[24000,44000,40000,20000,30000,450000]
year=[2015,2016,2017,2018,2019,2020]

plt.plot(year,price)
plt.ylim(10000,50000)
```

```python
# Grid
plt.plot(batsman['index'],batsman['V Kohli'],color='red',linestyle='solid')
plt.plot(batsman['index'],batsman['RG Sharma'],color='blue',linestyle='dashdot')

plt.title('Rohit Sharma Vs V. Kohli comparision')
plt.xlabel('Season')
plt.ylabel("Runs scores")
plt.grid()
```

```python
# show
plt.plot(batsman['index'],batsman['V Kohli'],color='red',linestyle='solid')
plt.plot(batsman['index'],batsman['RG Sharma'],color='blue',linestyle='dashdot')

plt.title('Rohit Sharma Vs V. Kohli comparision')
plt.xlabel('Season')
plt.ylabel("Runs scores")
plt.grid()
plt.show()
```

### Bar Chart
- Bivariate Analysis
- Numerical Vs Categorical
- Use Case: Aggregated analysis

```python
children=[10,20,30,40,50,60,70]
color=['red','blue','green','yellow','orange','pink','purple']
plt.bar(color,children)
```

```python
# Horizontal Bar chart
plt.barh(color,children)
```

```python
import pandas as pd
df = pd.read_csv('Datasets/batsman_season_record.csv')
df.head()
```

```python
# Secial Color every line
plt.bar(np.arange(df.shape[0])-0.2,df['2015'],width=0.2)
plt.bar(np.arange(df.shape[0]),df['2015'],width=0.2)
plt.bar(np.arange(df.shape[0])+0.2,df['2017'],width=0.2)
# plt.bar(np.arange(df.shape[0]+0.5),df['2016'],width=0.75)

plt.xticks(np.arange(df.shape[0]),df['batsman'])
plt.show()
```

### Pie Chart
- Univariate/Bivariate Analysis
- Categorical Or Numerical
- Use case - To find contribution 100 percent

```python
data=[10,40,50,30,22]
marks=['eng','science','math','sst','Hindi']
plt.pie(data,labels=marks)
plt.show()
```

```python
gayl=pd.read_csv('Datasets/gayle-175.csv')
gayl
```

```python
plt.pie(gayl['batsman_runs'],labels=gayl['batsman'],autopct='%0.1f%%')
plt.show()
```

```python
# Percentage and colors
plt.pie(gayl['batsman_runs'],labels=gayl['batsman'],autopct='%0.1f%%',colors=['blue','green','purple','yellow','pink'])
plt.show()
```

```python
plt.pie(gayl['batsman_runs'],labels=gayl['batsman'],autopct='%0.1f%%',explode=[0,0.1,0,0,0,0,],shadow=True)
plt.show()
```
