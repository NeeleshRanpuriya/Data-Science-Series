### Advance Matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Colord scatter plot

```python
iris = pd.read_csv('Datasets/iris.csv')
iris.sample(5)
```

```python
iris['Species']=iris['Species'].replace({'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2})
```

```python
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='viridis',alpha=1)
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar()
```

### Plot size

```python
plt.figure(figsize=(15,7))
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='viridis',alpha=1)
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar()
```

### Annotation

```python
batter=pd.read_csv('Datasets/batter.csv')
batter.sample(5)
```

```python
sample_df= batter.head(100).sample(25,random_state=5)
```

```python
sample_df.head()
```

```python
sample_df['avg'].values[0]
```

```python
plt.figure(figsize=(20,7))
plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs'])
plt.xlabel('Average')
plt.ylabel('Strike Rate')
plt.colorbar()

for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].iloc[i],sample_df['batter'].iloc[i])
```

```python
x=[1,2,3,4]
y=[6,7,8,9]

plt.scatter(x,y)
plt.text(1,6,'(1,6)',fontdict={'color':'red','size':15})
plt.text(2,7,'(2,7)')
plt.text(3,8,'(3,8)')
plt.text(4,9,'(4,9)')
```

### Horizontal or Vertical line

```python
plt.figure(figsize=(20,7))
plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs'])
plt.xlabel('Average')
plt.ylabel('Strike Rate')
plt.axhline(120,color='red',linestyle='--')
plt.axvline(30,color='red',linestyle='--')

for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].iloc[i],sample_df['batter'].iloc[i])
```

### 3D Surface plot

```python
x=np.linspace(-10,10,100)
y=np.linspace(-10,10,100)
```

```python
xx,yy=np.meshgrid(x,y)
```

```python
z=xx**2+yy**2
```

```python
fig=plt.figure(figsize=(10,8))
ax=fig.add_subplot(projection='3d')
col=fig.colorbar(ax.plot_surface(xx,yy,z,cmap='viridis',alpha=0.8))
fig.colorbar(col)
```

```python
z=np.sin(xx)*np.cos(yy)
fig=plt.figure(figsize=(10,8))
ax=fig.add_subplot(projection='3d')
col=fig.colorbar(ax.plot_surface(xx,yy,z,cmap='viridis',alpha=0.8))
fig.colorbar(col)
```

### Contour plot

```python
z=xx**2+yy**2
fig=plt.figure()
ax=fig.add_subplot()
ax.contour(xx,yy,z)
```

### Heatmap()

```python
ipl=pd.read_csv('Datasets/IPL_Ball_by_Ball_2008_2022.csv')
ipl.head()
temp_df=ipl[(ipl['ballnumber'].isin([1,2,3,4,5,6])) & (ipl['batsman_run']==6)]
grid=temp_df.pivot_table(index='overs',columns='ballnumber',values='batsman_run',aggfunc='count')
grid
```

```python
plt.figure(figsize=(20,10))
plt.imshow(grid,cmap='viridis')
plt.colorbar()
```
